```python
from __future__ import print_function

import pandas as pd
import numpy as np
%matplotlib inline
import keras as k
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras import activations
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.callbacks import Callback
```

```python
from keras.datasets import mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
"""Our neural network should have 60000 28*28 input. Which means we feed in 784 inputs flattened.
Inorder for neural network to work on this matrix, we choose to flatten the 2D vector into 1D vector,
hence we need to reshape the training and test vector."""
```

```python

cx_train=x_train
cx_test=x_test

vector=28*28
x_train=x_train.reshape(x_train.shape[0],vector)
x_train.shape
x_test=x_test.reshape(x_test.shape[0],vector)

all_train=np.concatenate((x_train,x_test), axis=0)
cx_train,cx_test
```

```python
#now change the output for 1 hot encoding way so that it is a vector of 10 units
"""Since, the output is the single integer digit ranging from 0-9, this if not optimized categorically will lead to
system behavior where largest digit are categorically more important than the small one.
Hence, we need to make a sparse binary matrix of 10 so that every output at that position is treated as 1 and rest of
them as 1.
eg= 3 is [0 0 0 1 0 0 0 0 0]"""
#cy_train=y_train
y_train=k.utils.to_categorical(y_train,10)
y_train.shape
#cy_test=y_test
y_test=k.utils.to_categorical(y_test,10)

all_test=np.concatenate((y_train,y_test), axis=0)
```

```python

#from tf.keras.callbacks import Callback
class NBatchLogger(Callback):
    def __init__(self, display):
        self.seen = 0
        self.display = display

    def on_batch_end(self, batch, logs={}):
        self.seen += logs.get('size', 0)
        if self.seen % self.display == 0:
            # you can access loss, accuracy in self.params['metrics']
            print('\n{}/{} - loss ....\n'.format(self.seen, self.params['metrics'][1])) 

out_batch = NBatchLogger(display=1)
```

```python
#Building models for recognizing digits: Part 1 - Sequential model

model=Sequential()
#adding layer
model.add(Dense(units=32,activation='sigmoid',input_shape=(784,)))
model.add(Dense(units=10))
model.add(Activation(tf.nn.softmax))
model.summary()

```

```python
#TRAINING

import keras.backend as K

def get_categorical_accuracy_keras(y_true, y_pred):
    return K.mean(K.equal(K.argmax(y_true, axis=1), K.argmax(y_pred, axis=1)))


from keras import optimizers
from keras import losses
sgd=optimizers.SGD(lr=0.5)
model.compile(loss=losses.categorical_crossentropy,optimizer=sgd,metrics=[get_categorical_accuracy_keras])
```

```python
#training only on train data and testing on validation data.
#history=model.fit(x_train,y_train,batch_size=100,epochs=17,verbose=0,callbacks=[out_batch])

history=model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=100,epochs=17,verbose=1)
```

```python
loss_val,metric_acc=model.evaluate(x_test,y_test,verbose=False)

plt.plot(history.history['get_categorical_accuracy_keras'])
plt.plot(history.history['val_get_categorical_accuracy_keras'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()
loss_val,metric_acc
```

```python
"""Multi-Layer Perceptron - 3 hidden layers - 1500 neurons each."""
#use truncated normal for weights
from keras import initializers
model_MLP=Sequential()
model_MLP.add(Dense(units=784,activation='relu',kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01, seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.1),input_shape=(784,)))
model_MLP.add(Dense(units=1500,activation='relu',kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01,seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.1)))
model_MLP.add(Dense(units=1500,activation='relu',kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01,seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.1)))
model_MLP.add(Dense(units=1500,activation='relu',kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01,seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.1)))
model_MLP.add(Dense(units=10,kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01,seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.1)))
model_MLP.add(Activation(tf.nn.softmax))
              
model_MLP.summary()

"""Biases are now -0.01."""
```

```python
adam=k.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=10e-8, decay=0.0, amsgrad=False)
model_MLP.compile(loss=losses.categorical_crossentropy,optimizer=adam,metrics=[get_categorical_accuracy_keras])
biases = model_MLP.layers[0].get_weights()[0]
"""Weight is not exceeding more than 2 STD. Check!"""
```

```python
 history=model_MLP.fit(x_train,y_train,validation_data=(x_test, y_test),batch_size=100,epochs=10,verbose=1)
```

```python
loss_val,metric_acc=model_MLP.evaluate(x_test,y_test,verbose=False)

plt.plot(history.history['get_categorical_accuracy_keras'])
plt.plot(history.history['val_get_categorical_accuracy_keras'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()
loss_val,metric_acc
```

```python
loss_val,metric_acc=model_MLP.evaluate(all_train,all_test,verbose=False)

plt.plot(history.history['get_categorical_accuracy_keras'])
plt.plot(history.history['val_get_categorical_accuracy_keras'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()
loss_val,metric_acc
```

# DROPOUT


```python
from keras.layers import Dropout

model_MLP=Sequential()
model_MLP.add(Dense(units=784,activation='relu',kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01, seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.1),input_shape=(784,)))
model_MLP.add(Dropout(0.5))
model_MLP.add(Dense(units=1500,activation='relu',kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01,seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.1)))
model_MLP.add(Dropout(0.5))
model_MLP.add(Dense(units=1500,activation='relu',kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01,seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.1)))
model_MLP.add(Dropout(0.5))
model_MLP.add(Dense(units=1500,activation='relu',kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01,seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.1)))
model_MLP.add(Dense(units=10,kernel_initializer=initializers.TruncatedNormal(mean=0.0, stddev=0.01,seed=None),use_bias=True,bias_initializer=initializers.Constant(value=-0.0)))
model_MLP.add(Activation(tf.nn.softmax))
              
model_MLP.summary()
```

```python
adam=k.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=10e-8, decay=0.0, amsgrad=False)
model_MLP.compile(loss=losses.categorical_crossentropy,optimizer=adam,metrics=[get_categorical_accuracy_keras])
```

```python
history=model_MLP.fit(x_train,y_train,validation_data=(x_test, y_test),batch_size=100,epochs=10,verbose=1)
```

```python
loss_val,metric_acc=model_MLP.evaluate(x_test,y_test,verbose=False)
plt.plot(history.history['get_categorical_accuracy_keras'])
plt.plot(history.history['val_get_categorical_accuracy_keras'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()
loss_val,metric_acc
```

# CNN

```python
x_train = cx_train.reshape(60000,28,28,1)
x_test = cx_test.reshape(10000,28,28,1)
```

```python
from keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D, GlobalAveragePooling2D
from keras.layers import Dense, Dropout, Activation, Flatten
model = Sequential()    
model.add(Conv2D(32, kernel_size=(5, 5),strides=(1, 1),input_shape=(28, 28,1), use_bias=True,
                 bias_initializer=initializers.Constant(value=-0.1),activation='relu',padding='same'))    
model.add(MaxPooling2D(pool_size=(2, 2),strides=(2,2)))    

model.add(Conv2D(64, kernel_size=(5, 5),strides=(1, 1),use_bias=True,
                 bias_initializer=initializers.Constant(value=-0.1),activation='relu',padding='same'))    
model.add(MaxPooling2D(pool_size=(2, 2),strides=(2,2)))
model.add(Flatten())    
model.add(Dense(10,use_bias=True,
                 bias_initializer=initializers.Constant(value=-0.1)))
model.add(Activation(tf.nn.softmax))
model.summary()
```

```python
adam=k.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=10e-8, decay=0.0, amsgrad=False)
model.compile(loss=losses.categorical_crossentropy,optimizer=adam,metrics=[get_categorical_accuracy_keras])
```

```python
history=model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=100, verbose=1)
```

```python
loss_val,metric_acc=model.evaluate(x_test,y_test,verbose=False)
plt.plot(history.history['get_categorical_accuracy_keras'])
plt.plot(history.history['val_get_categorical_accuracy_keras'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()
loss_val,metric_acc
```
