```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#uploading training data file ( x1,x2,obs)
train=pd.read_csv('TrainingRidge.csv')
train_x=train[['x1','x2']]
train_y=train['obs']

#validation file
val=pd.read_csv('ValidationRidge-Y.csv')
val_x=val[['x1','x2']]
val_y=val['dens']

```
